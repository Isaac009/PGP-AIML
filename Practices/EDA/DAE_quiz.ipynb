### n4. Which of the following can be used to see the frequency distribution of 'origin'?

- sns.countplot( ) is used to display the counts of observations in each categorical bin using bars.
- sns.histplot( ) is used to display the distribution plot of quantitative variables.
- sns.jointplot( ) is used to display a plot of two variables with bivariate and univariate graphs.
- sns.catplot( ) is used to display multiple categorical plots onto a FacetGrid
 

Please note that we can not directly use histograms while plotting categorical data. We know that the bars in the count plot are separated whereas the histograms always show adjacent bars. Since we use continuous data for histograms, the x-axis of the histogram indeed suggests continuity and that's why we get adjacent bars. If we use histograms for plotting categorical data that will be misleading because there is no continuity in the categorical data. Therefore, a histogram can not be used to show the frequency distribution of a categorical variable.
